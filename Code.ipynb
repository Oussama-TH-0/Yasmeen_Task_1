```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv("/kaggle/input/employee-data/Employee 1000x.csv")
print(df.shape)
```

```python
# file contains data for 10000 employee with 8 columns 
df.head(3) # to print the data of the first 3 employees
```

```python
# check if we have empty cells of Null Values
null_data = df.isnull().values.any()
print(null_data)
```

```python
#since we dont have any missing infomation lets start the analysis
df["Sex"].value_counts() / df.shape[0] *100
```

```python
title_count = df['Job Title'].value_counts().reset_index()
title_count.columns = ['Job Title', 'Count']
title_count
```

```python
# to visualise the titles in chart we need some libraries
import seaborn as sb
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sb.barplot(data=title_count.head(20),x="Job Title",y="Count", palette='viridis')
plt.xticks(rotation=90)
plt.title('Number of Employees by Job Title')
plt.tight_layout()
plt.show()
```
